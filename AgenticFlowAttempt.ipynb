```python
from crewai import Agent, Task, Crew, Process
from langchain.chat_models import ChatOpenAI
import os

os.environ["OPENAI_MODEL_NAME"] = "gpt-3.5-turbo"
openai_api_key = "your_openai_api_key_here"
llm = ChatOpenAI(openai_api_key=openai_api_key, temperature=0)

file="/content/graph.txt"
with open(file, 'r') as f:
    text = f.read()

graphic=Agent(
    role='graphic designer',
    backstory='you are an expert graphic designer',
    goal=f'improve the text to generate a better graph',
    allow_delegation=False,
    llm=llm,
    verbose=True
)

graphic_task=Task(
    description=f'improve the given text: {text}',
    expected_output='a file of triples',
    agent=graphic
)

writer=Agent(
    role='writer',
    backstory='you are a writer',
    goal='write a file',
    allow_delegation=False,
    llm=llm,
    verbose=True
)

writer_task=Task(
    description='write a file',
    expected_output='a file with a triple per line, every element must be separated by: | ',
    agent=writer
)

crew= Crew(
    agents=[graphic, writer],
    tasks=[graphic_task, writer_task],
    process=Process.hierarchical,
    verbose=True,
    manager_llm=llm
)

result=crew.kickoff()

with open('/content/refined_graph.html', 'w') as file:
    file.write(result)
```
